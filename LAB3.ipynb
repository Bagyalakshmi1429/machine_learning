ASSOCIATION RULES MINING



```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/Market_Basket_Optimisation (1).csv')

# Convert the transaction data into a list of lists (each list represents a transaction)
transactions = []
for idx, row in df.iterrows():
    transactions.append(row.dropna().tolist())

# Initialize a TransactionEncoder
te = TransactionEncoder()
# Transform the transaction data using the TransactionEncoder
oht_array = te.fit(transactions).transform(transactions)
oht_df = pd.DataFrame(oht_array, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(oht_df, min_support=0.05, use_colnames=True)

# Generate association rules
association_rules_df = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Sort association rules by lift in descending order
sorted_rules = association_rules_df.sort_values(by='lift', ascending=False)

# Print the top association rules
num_top_rules = 10
print("Top Association Rules:")
print(sorted_rules.head(num_top_rules))
```

INTERPRETAION OF ASSOCIATION RULES

```python

for index, rule in association_rules_df.iterrows():
    antecedents = list(rule['antecedents'])
    consequents = list(rule['consequents'])
    support = rule['support']
    confidence = rule['confidence']
    lift = rule['lift']

    print(f"Rule: {antecedents} -> {consequents}")
    print(f"Support: {support:.3f}")
    print(f"Confidence: {confidence:.3f}")
    print(f"Lift: {lift:.3f}")
    print("\n")
```

VISUALIZE ASSOCIATION RULES

```python


heatmap_data = association_rules_df.pivot(index='antecedents', columns='consequents', values='confidence')
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Association Rule Confidence Heatmap')
plt.xlabel('Consequents')
plt.ylabel('Antecedents')

plt.show()
```
